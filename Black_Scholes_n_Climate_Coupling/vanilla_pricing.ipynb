```python
import pandas as pd
import glob
import os
import math
from scipy.stats import norm

files = sorted(glob.glob("1*.csv"))

data = []
#counter = 0
for file in files:
    
    # Extract company name from filename
    company_name = os.path.basename(file).replace(".csv", "")
    current_data = pd.read_csv(file)
    current_data['Return'] = 0.0 #Set new column "Return" value to 0
    
    for j in range(1, len(current_data)): #Loop through each row in the csv file
        closing_ratio = ((current_data['Close']).iloc[j])/((current_data['Close']).iloc[j-1]) #Closing Price Ratio: Close(n)/Close(n-1)
        current_data.loc[j, 'Return'] = math.log(closing_ratio) #Calculate Log Returns
        
    sigma = ((current_data['Return']).std())*math.sqrt(252) #Calculate standard deviation of returns ac
    S0 = current_data['Close'].iloc[-1]
    K = S0 # (vary)
    r = 0.035 # (vary)
    T = 1 # (vary)
        #   from: https://www.cnbc.com/quotes/US1Y
    d1 = (((r + 1/2*sigma**2)*T) + math.log(S0/K))/(sigma*math.sqrt(T))
    d2 = d1 - sigma*(math.sqrt(T))
    option_price = S0*(norm.cdf(d1))-K*(math.exp(-r*T))*(norm.cdf(d2))
    print(company_name[1:] + ": " + str(option_price))
        #print(current_data['Return'])
    

```
