# No-outliers Absolute % Change Bar Plot

This notebook loads `artifacts/summary_no_outliers.csv` and plots mean absolute % change by company with Â±1 std error bars.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt

ART_DIR = 'artifacts'
summary_fp = os.path.join(ART_DIR, 'summary_no_outliers.csv')
summary = pd.read_csv(summary_fp)
summary_plot = summary.sort_values('mean_abs_pct', ascending=True)

plt.figure(figsize=(12, max(5, 0.35 * len(summary_plot))))
plt.barh(summary_plot['company'], summary_plot['mean_abs_pct'], xerr=summary_plot['std_abs_pct'],
         alpha=0.85, color='seagreen')
plt.xlabel('Mean absolute % change (no-outliers; error bar = 1 std)')
plt.ylabel('Company')
plt.title('Mean absolute % change by company (no outliers removed)')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

out_fp = os.path.join(ART_DIR, 'barplot_mean_abs_pct_no_outliers.png')
plt.savefig(out_fp, dpi=200, bbox_inches='tight')
print('Saved:', out_fp)
```
